```python
import requests
import pandas as pd

url = "https://openapi.twse.com.tw/v1/opendata/t187ap03_L"
data = requests.get(url).json()
df_twse = pd.DataFrame(data)[["公司代號", "公司名稱", "產業別"]]
df_twse.columns = ["stock_id", "name", "sector"]
df_twse["stock_id"] = df_twse["stock_id"].astype(str)
df_twse.head()
```

```python
df_twse
```

```python
df = df.merge(df_twse[["stock_id", "sector"]], on="stock_id", how="left")
```

```python
df.to_parquet('../data/twse_2020_to_2025_sector.parquet')
```
